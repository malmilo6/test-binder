# Programming Assignment: Multi-Format Data Loading

Welcome to the assignment for Multi-Format Data Loading.

Data is a raw or basic collection of observations, measurements, facts etc. that is used to represent and describe objects, phenomena and events. On the other hand, datasets are organized pieces of data, that are typically related to a specific domain or have some common logic behind their organization. Data and datasets provide probably the most important and crucial details without which machine learning could not evolve so much over the years. One of the most important things about high-quality data is that it provides a way for machine learning models to "learn", or find underlying patterns and relations between various pieces of data.

Datasets can originate from various sources, such as: experiments, surveys, systems logs, sensors, databases and many others. They vary in size and format. Usually, datasets are organized based on structurality. By this criterion, datasets can be:
1. **Structured** - highly organized/formatted data represented in strict predefined format (with a strict "schema"). Usually, this type of data is represented in Relational Databases, CSV/Excel files and others. Below is represented an example of a dataset organized in tabular format:

    | **ID** | **Age** | **Body_Mass_Index** | **Gender** | **Blood_Type** | **...** | **Is_Sick** |
    |:------:|:-------:|:-------------------:|:----------:|:--------------:|:-------:|:-----------:|
    |   _1_  |    18   |          40         |     "M"    |      "AB+"     |   ...   |     True    |
    |   _2_  |    35   |          27         |    None    |       "O"      |   ...   |    False    |
    |   _3_  |    24   |          31         |     "F"    |      None      |   ...   |    False    |
    |  _..._ |   ...   |         ...         |     ...    |       ...      |   ...   |     ...     |
    | _1000_ |    95   |          35         |     "F"    |      "A-"      |   ...   |     True    |

2. **Semi-Structured** - data that have some organizational properties, with flexible "schema", that may differ from dataset record to record. Most of the time, this type of datasets is represented in JSON, XML, system logs and other formats. Below is represented an example of such JSON dataset (Note differences in these records):

    ```json
    {
        "name": "John",
        "age": 20,
        "interests": ["Literature", "Music Production"],
        "num_requests": 201,
        "is_subscribed": "True"
    },
    {
        "name": "Jane",
        "age": "None",
        "interests": ["Literature", "Music Production"],
        "is_subscribed": "False",
        "location": "Europe"
    },
    ...
    ```

3. **Unstructured** - data that contain information that is does not conform to traditional data models, with no "schema". This type of datasets can be represented in various ways: videos/audio/text files, chat messages, etc. This category is the hardest to work with, but most of the time, they capture rich insights and information. 

Most of Machine Learning application involve working with structured datasets, however datasets of other two types are also used, but require more preparation and deeper understanding of complex loading and processing mechanisms. For the start, being able to work with datasets involve being able to load them properly into data structures that can be manipulated/actioned on via programming. One of the most popular libraries that offer this possibility is [`pandas`](https://pandas.pydata.org/docs/index.html). [`pandas`](https://pandas.pydata.org/docs/index.html) is an open-source data analysis and manipulation tool, built on top of `Python` programming language, that is able to convert data from files into data structures that can be interacted with via code. This library offers two main data structures that can be used to represent tabular data:
1. `Series` - one-dimensional labeled array holding data of any type. This structure can be interpreted as a basic list. For example:

    | **ID** |
    |:------:|
    |   _1_  |
    |   _2_  |

    | _1000_ |

2. `DataFrame` - two-dimensional data structure that holds data like a two-dimensional array. In other words, this is a basic table with rows and columns. For example, recall previous structured dataset example:

    | **ID** | **Age** | **Body_Mass_Index** |
    |:------:|:-------:|:-------------------:|
    |   _1_  |    18   |          40         |
    |   _2_  |    35   |          27         |
    |   _3_  |    24   |          31         |
    |  _..._ |   ...   |         ...         |

### 1.1 Introduction to `pd.Series`
Series are formed of the following key elements:

1. **Data** – A one-dimensional array of values.
2. **Index** – Labels that identify each element in the Series.
3. **Name** - Name of the Series object.
4. **Dtype** – The data type of the values (e.g., `int64`, `float64`, `object`).

These structures can be instantiated in the following ways:

1. From a list or tuple
2. From a NumPy array
3. From a dictionary
4. From a scalar value (with a specified index)
5. From another Series or DataFrame column

````python
# 1.1 Using a list
list_of_data: list[Any] = [1, "a", 3.2, [1, 2, 3]]

series_from_list: pd.Series = pd.Series(data=list_of_data, name='Sample_Data')
series_from_list
````

````python
# 1.2 Using a tuple
tuple_of_data: tuple[Any] = (1, "a", 3.2, [1, 2, 3])

series_from_tuple: pd.Series = pd.Series(data=tuple_of_data, name='Sample_Data')
series_from_tuple
````

````python
# 2. Using a numpy 1D array
one_d_np_array: np.ndarray = np.array(object=[1, 2, 3, 4, 5])

series_from_np_array: pd.Series = pd.Series(data=one_d_np_array, name="Sample_Data")
series_from_np_array
````

````python
# 3. Using a dictionary of values
# Dict keys represent indices of the Series
# Dict values represent values that are stored in the Series at their index
dict_of_data: dict[str, Any] = {
    "Index_A": [1, "A", 3.2],
    "Index_B": 1,
    "Index_C": "B"
}

series_from_dict: pd.Series = pd.Series(data=dict_of_data, name='Sample_Data')
series_from_dict
````

````python
# 4. Using a scalar with index
# In this case, a single value can be propagated at more indices
value_to_propagate: Any = "VALUE"
length_indices: list[Any] = ["A", 1, "ANY"]

series_from_scalar: pd.Series = pd.Series(data=value_to_propagate, index=length_indices)
series_from_scalar
````

````python
# 5.1 Using another Series
series_orig: pd.Series = pd.Series(data=[10, 20, 30], name="C")

series_from_anth_series: pd.Series = pd.Series(data=series_orig) # or `series_orig.copy(deep=True/False)`
series_from_anth_series
````

````python
# 5.2 Using a column from DataFrame
df_1: pd.DataFrame = pd.DataFrame(data=[[10, 20, 30], [20, 30, 40], [30, 40, 50]], columns=["Feature_1", "Feature_2", "Feature_3"])

series_from_anth_series: pd.Series = pd.Series(data=df_1["Feature_1"])  # You can access DataFrame by column names similar to list index access = df_1["Feature_1"]
series_from_anth_series
````

````python
# You may also rename a specific index label/name using `rename()` method of `pd.Series`
series_orig: pd.Series = pd.Series(data=[10, 20, 30], index=["INDEX_1", "INDEX_2", "INDEX_3"], name="C")

new_series: pd.Series = series_orig.rename({"INDEX_1": "NEW_INDEX_NAME"})
new_series
````

### 1.2 Introduction to `pd.DataFrame`
DataFrames are formed of the following key elements:

1. **Data** – The actual values stored in tabular form (rows and columns).
2. **Index** – The row labels that identify each row uniquely.
3. **Columns** – The column labels (names) that identify each variable/field.
4. **Dtypes** – The data types of each column (e.g., int, float, string, bool).

These structures can be instantiated in various ways:

1. From a dictionary of lists or arrays
2. From a list of dictionaries
3. From a NumPy array
4. From another DataFrame or Series
5. From external files (CSV, Excel, SQL, JSON, etc.) - will be covered further in the notebook.

````python
# 1. Using a dictionary with values
# keys = column names
# values = values associated with a specific key/column
dict_of_data: dict[str, Any] = {
    "feature_1": [1.2, 2, 3],
    "feature_2": [3, 4.2, 5],
    "feature_3": [5, 5.3, 1.2]
}
df_from_dict: pd.DataFrame = pd.DataFrame(data=dict_of_data)
df_from_dict
````

````python
# 2. Using a list of dictionaries
# Each key-value pair in the dictionaries represent a value associated with a column/key
list_of_dicts: list[dict] = [
    {"A": 1, "B": 4},
    {"A": 2, "B": 5},
    {"A": 3, "C": 3}
]
df_from_list: pd.DataFrame = pd.DataFrame(data=list_of_dicts)
df_from_list
````

````python
# 3. Using numpy arrays
# In this case, the contents of the numpy array are treated as values
# By default, it will use numbers for columns, unless you specify the actual names for columns
#   inside the argument for pd.DataFrame
np_array: np.ndarray = np.array([[1, 2, 3], [3, 4, 5]])

df_from_np_array = pd.DataFrame(data=np_array, columns=["A", "B", "C"])
df_from_np_array
````

````python
# 4.1 From another Dataframe (copy)
df_original = pd.DataFrame(
    data={
        "A": [1, 2, 3],
        "B": [4, 5, 6],
        "C": [7, 8, 9]
    }
)

df_copy = pd.DataFrame(data=df_original)
df_copy
````

````python
# 4.2 From a subset from original DataFrame
df_subset = pd.DataFrame(data=df_original[["A", "B"]])
df_subset
````

````python
# 4.3 From Series
series = pd.Series(data=[10, 20, 30], name="C")

df_from_series = pd.DataFrame(data=series)
df_from_series
````

````python
# 4.4 From multiple Series as separate columns
series_1: pd.Series = pd.Series(data=[10, 20, 30], name="C")
series_2: pd.Series = pd.Series(data=["A", "B", "C"], name="D")
series_3: pd.Series = pd.Series(data=[4.3, "ABC", 3.211], name="E")

col_name_values_map: dict[str, pd.Series] = {
    # Use for column names Series name
    series_1.name: series_1,
    # Use custom name as column name
    "Another_Name": series_2,
    # or other data type
    3: series_3
}

df_from_series = pd.DataFrame(data=col_name_values_map)
df_from_series
````

## 4 - Loading from JSON file

Another common format for public datasets used in ML applications is **JSON** (JavaScript Object Notation) files. JSON is a lightweight, text-based format for representing structured data. It is widely used for data interchange between applications and APIs due to its human-readable syntax and ease of parsing by machines. Unlike tabular formats like `.csv` or spreadsheet formats like `.xlsx`, JSON files can represent hierarchical data, such as nested objects or arrays, making them very flexible for complex datasets.

JSON datasets are commonly structured as:
1. **Objects** - collections of key-value pairs enclosed in `{}`.
2. **Arrays** - ordered lists of values enclosed in `[]`.
3. **Nested structures** - objects or arrays can be nested inside other objects or arrays to represent relationships between data points.

Because JSON can represent deeply nested or irregular data, it can be slightly more challenging to flatten into standard tabular formats, but it provides a very flexible way to store and transport complex datasets.

Python provides multiple ways to work with JSON files:
- The built-in `json` module can be used to read - `json.load()` or write - `json.dump()` JSON objects. In both cases, classical `python` objects will be used - dictionaries, lists, etc.
- `pandas` provides the `pd.read_json()` function, which can load a JSON file or JSON-formatted string directly into a `pd.DataFrame`. Under the hood, `pandas` handles different orientations (records, columns, index) and can parse nested structures into flat tables.

## 5 - Loading from SQL Database

Another widely used source of public datasets in ML and data analysis workflows are **SQL databases**. SQL (Structured Query Language) databases store data in a structured, tabular format consisting of rows and columns, organized into tables that can be related to each other through keys. Unlike flat files such as `.csv` or hierarchical formats like JSON, SQL databases are designed for efficient querying, filtering, aggregation, and joining of large volumes of structured data.

A SQL dataset is typically stored inside a **database**, which contains:
1. **Tables** – structured collections of rows and columns (similar to DataFrames).
2. **Schemas** – logical groupings of related tables and metadata.
3. **Relations** – links between tables defined using primary and foreign keys.
4. **Indexes** – data structures that speed up search and retrieval operations.

Because SQL databases enforce schemas and data types, they provide strong guarantees about data consistency and integrity. They are commonly used in production systems, analytics pipelines, and large-scale ML workflows where datasets are too large or dynamic to be stored as static files.

In SQL databases, data is queried using SQL statements such as:
```sql
SELECT sepal_length, sepal_width, species
FROM iris
WHERE species = 'setosa';
```
This query retrieves a subset of rows and columns, from a table named `iris`, that belong to target class `setosa`. Such queries allow efficient filtering and aggregation directly within the database engine before loading the results into Python.

Python provides several ways to interact with SQL datasets:

1. The built-in sqlite3 module can be used to execute SQL queries against SQLite databases.
2. Database-specific connectors (e.g., psycopg2 for PostgreSQL, mysql-connector for MySQL) allow communication with remote databases.
3. pandas provides high-level functions such as `pd.read_sql()`, `pd.read_sql_query()`, and `pd.read_sql_table()` to load SQL query results directly into a `pd.DataFrame`.

Below is a small conceptual example of a SQL table representing a subset of the Iris dataset:

| id  | sepal_length | sepal_width | petal_length | petal_width |  species  |
| :-: | :----------: | :---------: | :----------: | :---------: | :-------: |
|  1  |      5.1     |     3.5     |      1.4     |     0.2     |   setosa  |
|  2  |      4.9     |     3.0     |      1.4     |     0.2     |   setosa  |
|  3  |      6.2     |     3.4     |      5.4     |     2.3     | virginica |

Using `pd.read_sql()`, you can execute a SQL query and load the returned result set directly into a `pd.DataFrame`, combining the flexibility of SQL querying with the convenience of pandas data manipulation tools.

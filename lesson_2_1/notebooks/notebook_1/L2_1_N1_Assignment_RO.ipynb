# Temă de Programare: Încărcarea Datelor utilizând Multiple Formate

Bun venit la tema dedicată Încărcării Datelor utilizând Multiple Formate.

Datele reprezintă o colecție brută sau de bază de observații, măsurători, fapte etc., utilizate pentru a reprezenta și descrie obiecte, fenomene și evenimente. Pe de altă parte, seturile de date sunt colecții organizate de date, de obicei legate de un domeniu specific sau cu o logică comună în organizarea lor. Datele și seturile de date oferă probabil cele mai importante și esențiale detalii, fără de care machine learning-ul nu ar fi putut evolua atât de mult de-a lungul anilor. Unul dintre cele mai importante aspecte ale datelor de înaltă calitate este că oferă posibilitatea modelelor de machine learning de a "învăța" sau de a descoperi tipare și relații subiacente între diverse date.

Seturile de date pot proveni din diverse surse, cum ar fi: experimente, sondaje, jurnale de sistem, senzori, baze de date și multe altele. Ele variază ca dimensiune și format. De obicei, seturile de date sunt organizate în funcție de structuralitate. După acest criteriu, seturile de date pot fi:
1. **Structurate** - date foarte organizate/formatate, reprezentate într-un format predefinit strict (cu o "schemă" strictă). De obicei, acest tip de date este reprezentat în baze de date relaționale, fișiere CSV/Excel și altele. Mai jos este reprezentat un exemplu de set de date organizat în format tabelar:

    | **ID** | **Vârstă** | **Indice_Masă_Corporală** | **Gen** | **Grup_Sangvin** | **...** | **Este_Bolnav** |
    |:------:|:----------:|:-------------------------:|:-------:|:----------------:|:-------:|:---------------:|
    |   _1_  |    18      |          40               |   "M"   |      "AB+"        |   ...   |     True        |
    |   _2_  |    35      |          27               |   None  |       "O"         |   ...   |    False        |
    |   _3_  |    24      |          31               |   "F"   |      None         |   ...   |    False        |
    |  _..._ |   ...      |         ...               |   ...   |       ...         |   ...   |     ...         |
    | _1000_ |    95      |          35               |   "F"   |      "A-"         |   ...   |     True        |

2. **Semi-Structurate** - date care au unele proprietăți organizatorice, cu o "schemă" flexibilă, ce poate diferi de la o înregistrare la alta. De cele mai multe ori, acest tip de seturi de date este reprezentat în JSON, XML, jurnale de sistem și alte formate. Mai jos este reprezentat un exemplu de astfel de set de date JSON (observați diferențele dintre înregistrări):

    ```json
    {
        "name": "John",
        "age": 20,
        "interests": ["Literatură", "Producție Muzicală"],
        "num_requests": 201,
        "is_subscribed": "True"
    },
    {
        "name": "Jane",
        "age": "None",
        "interests": ["Literatură", "Producție Muzicală"],
        "is_subscribed": "False",
        "location": "Europa"
    },
    ...
    ```

3. **Nestructurate** - date care conțin informații ce nu se conformează modelelor tradiționale de date, fără o "schemă". Acest tip de seturi de date poate fi reprezentat în diverse moduri: fișiere video/audio/text, mesaje de chat etc. Această categorie este cea mai dificil de utilizat, dar de cele mai multe ori capturează informații și perspective bogate.

Majoritatea aplicațiilor de Machine Learning implică lucrul cu seturi de date structurate, însă și celelalte două tipuri sunt utilizate, dar necesită mai multă pregătire și o înțelegere mai profundă a mecanismelor complexe de încărcare și procesare. La început, capacitatea de a lucra cu seturi de date implică abilitatea de a le încărca corect în structuri de date care pot fi manipulate/acționate prin programare. Una dintre cele mai populare biblioteci care oferă această posibilitate este [`pandas`](https://pandas.pydata.org/docs/index.html). [`pandas`](https://pandas.pydata.org/docs/index.html) este un instrument open-source de analiză și manipulare a datelor, construit pe baza limbajului de programare `Python`, capabil să convertească date din fișiere în structuri de date cu care se poate interacționa prin cod. Această bibliotecă oferă două structuri principale de date ce pot fi utilizate pentru a reprezenta date tabulare:
1. `Series` - un array etichetat unidimensional ce poate conține date de orice tip. Această structură poate fi interpretată ca o listă de bază. De exemplu:

    | **ID** |
    |:------:|
    |   _1_  |
    |   _2_  |
    |   _3_  |
    |  _..._ |
    | _1000_ |

2. `DataFrame` - o structură de date bidimensională ce conține date precum un array bidimensional. Cu alte cuvinte, aceasta este un tabel de bază cu rânduri și coloane. De exemplu, reamintim exemplul anterior de set de date structurat:

    | **ID** | **Vârstă** | **Indice_Masă_Corporală** |
    |:------:|:----------:|:-------------------------:|
    |   _1_  |    18      |          40               |
    |   _2_  |    35      |          27               |
    |   _3_  |    24      |          31               |
    |  _..._ |   ...      |         ...               |
    | _1000_ |    95      |          35               |

Biblioteca `pandas` oferă două structuri principale de date - `pd.DataFrame` și `pd.Series`. Ambele structuri pot conține orice valori, la fel ca listele Python, dar sunt compatibile cu alte biblioteci, cum ar fi `numpy`, ceea ce înseamnă că la instanțierea obiectelor `pd.DataFrame` și `pd.Series`, ca date de intrare pot fi folosite array-uri `numpy`. ### 1.1 Introducere în `pd.Series`
Un Series este format din următoarele elemente cheie:

1. **Date** – Un array unidimensional de valori.
2. **Index** – Etichete care identifică fiecare element din Series.
3. **Nume** - Numele obiectului Series.
3. **Dtype** – Tipul de date al valorilor (ex: `int64`, `float64`, `object`).

Aceste structuri pot fi instanțiate în următoarele moduri:

1. Dintr-o listă sau tuplu
2. Dintr-un array NumPy
3. Dintr-un dicționar
4. Dintr-o valoare scalară (cu un index specificat)
5. Dintr-un alt Series sau dintr-o coloană DataFrame


````python
# 1.1 Folosind o listă
list_of_data: list[Any] = [1, "a", 3.2, [1, 2, 3]]

series_from_list: pd.Series = pd.Series(data=list_of_data, name='Sample_Data')
series_from_list
````

````python
# 1.2 Folosind un tuplu
tuple_of_data: tuple[Any] = (1, "a", 3.2, [1, 2, 3])

series_from_tuple: pd.Series = pd.Series(data=tuple_of_data, name='Sample_Data')
series_from_tuple
````

````python
# 2. Folosind un array numpy 1D
one_d_np_array: np.ndarray = np.array(object=[1, 2, 3, 4, 5])

series_from_np_array: pd.Series = pd.Series(data=one_d_np_array, name="Sample_Data")
series_from_np_array
````

````python
# 3. Folosind un dicționar de valori
# Cheile dicționarului reprezintă indicii din Series
# Valorile dicționarului reprezintă valorile stocate în Series la indexul lor
dict_of_data: dict[str, Any] = {
    "Index_A": [1, "A", 3.2],
    "Index_B": 1,
    "Index_C": "B"
}

series_from_dict: pd.Series = pd.Series(data=dict_of_data, name='Sample_Data')
series_from_dict
````

````python
# 4. Folosind o valoare scalară cu index
# În acest caz, o singură valoare poate fi propagată la mai mulți indici
value_to_propagate: Any = "VALUE"
length_indices: list[Any] = ["A", 1, "ANY"]

series_from_scalar: pd.Series = pd.Series(data=value_to_propagate, index=length_indices)
series_from_scalar
````

````python
# 5.1 Folosind un alt Series
series_orig: pd.Series = pd.Series(data=[10, 20, 30], name="C")

series_from_anth_series: pd.Series = pd.Series(data=series_orig) # sau `series_orig.copy(deep=True/False)`
series_from_anth_series
````

````python
# 5.2 Folosind o coloană dintr-un DataFrame
df_1: pd.DataFrame = pd.DataFrame(data=[[10, 20, 30], [20, 30, 40], [30, 40, 50]], columns=["Feature_1", "Feature_2", "Feature_3"])

series_from_anth_series: pd.Series = pd.Series(data=df_1["Feature_1"])  # Poți accesa DataFrame după denumirile coloanelor similar cu accesul prin indexul listei = df_1["Feature_1"]
series_from_anth_series
````

````python
# Poți, de asemenea, redenumi o etichetă/denumire de index specifică folosind metoda `rename()` a `pd.Series`
series_orig: pd.Series = pd.Series(data=[10, 20, 30], index=["INDEX_1", "INDEX_2", "INDEX_3"], name="C")

new_series: pd.Series = series_orig.rename({"INDEX_1": "NEW_INDEX_NAME"})
new_series
````

1. **Date** – Valorile actuale stocate în formă tabelară (rânduri și coloane).

3. **Coloane** – Etichetele coloanelor (denumirile) care identifică fiecare variabilă/câmp.
4. **Dtype-uri** – Tipurile de date ale fiecărei coloane (ex: int, float, string, bool).

Aceste structuri pot fi instanțiate în diverse moduri:

1. Dintr-un dicționar de liste sau array-uri
2. Dintr-o listă de dicționare

Un alt format frecvent al seturilor de date publice utilizate pentru aplicațiile ML îl reprezintă fișierele `.xls` și `.xlsx`, sau pur și simplu fișierele Excel. Acestea sunt de obicei create și editate folosind software de foi de calcul (ex.: Microsoft Excel, Google Spreadsheets etc.). Spre deosebire de formatele text simplu (ex.: `.csv`), fișierele Excel pot conține mai multe foi de lucru într-un singur fișier, permițând organizarea seturilor de date pe diferite tab-uri. Seturile de date Excel suportă caracteristici avansate, cum ar fi: formatarea celulelor, formule, grafice și celule îmbinate, care pot fi utile pentru analiza și prezentarea datelor. Totuși, aceste caracteristici suplimentare fac fișierele Excel mai puțin ușoare și mai puțin lizibile de către om comparativ cu fișierele `.csv`.

Sunt mai dificil de deserializat în obiecte python datorită varietății ridicate de formate de serializare care pot fi folosite:
1. `.xlsx` - format XML
2. `.xlsb` - format Binar
3. `.xlsm` - cu suport pentru Macro-uri
4. `.ods` - format standardizat open-source similar cu formatul Microsoft Excel (ex.: folosit în LibreOffice)
5. și chiar în format `.csv`

Luând în considerare toate acestea, fișierele Excel sunt mai dificil de operat și de încărcat în obiecte python. Există mai multe biblioteci care fac asta, cum ar fi: `pandas` - cu funcția `pd.read_excel()`, care încarcă conținutul fișierului Excel direct într-un `pd.DataFrame` (cu care ești deja familiar), `openpyxl` - utilizată pentru citirea fișierelor `.xlsx`/`.xlsm`, `xlrd` - pentru citirea fișierelor Excel legacy `.xls`, `pyxlsb` - folosită pentru citirea fișierelor Excel binare `.xlsb` și altele. Sub acoperire, `pandas` utilizează toate bibliotecile menționate anterior, cât și altele, pentru a gestiona diferite tipuri de fișiere Excel.

## 4 - Încărcarea din fișier JSON


Un alt format frecvent pentru seturile de date publice utilizate în aplicațiile ML îl reprezintă fișierele **JSON** (JavaScript Object Notation). JSON este un format ușor, bazat pe text, pentru reprezentarea datelor structurate. Este utilizat pe scară largă pentru schimbul de date între aplicații și API-uri datorită sintaxei sale lizibile și ușurinței de parsare de către mașini. Spre deosebire de formatele tabulare precum `.csv` sau formatele de foi de calcul precum `.xlsx`, fișierele JSON pot reprezenta date ierarhice, cum ar fi obiecte sau array-uri imbricate, făcându-le foarte flexibile pentru seturi de date complexe.

Seturile de date JSON sunt de obicei structurate ca:
1. **Obiecte** - colecții de perechi cheie-valoare incluse în `{}`.
2. **Array-uri** - liste ordonate de valori incluse în `[]`.
3. **Structuri imbricate** - obiectele sau array-urile pot fi imbricate în alte obiecte sau array-uri pentru a reprezenta relațiile dintre punctele de date.

Deoarece JSON poate reprezenta date profund imbricate sau neregulate, poate fi ușor mai dificil de aplatizat în formate tabulare standard, dar oferă o modalitate foarte flexibilă de a stoca și transporta seturi de date complexe.

Python oferă mai multe modalități de a lucra cu fișiere JSON:
- Modulul built-in `json` poate fi folosit pentru citire - `json.load()` sau scriere - `json.dump()` a obiectelor JSON. În ambele cazuri, vor fi utilizate obiecte clasice `python` - dicționare, liste etc.
- `pandas` oferă funcția `pd.read_json()`, care poate încărca un fișier JSON sau un string formatat JSON direct într-un `pd.DataFrame`. Sub acoperire, `pandas` gestionează diferite orientări (records, columns, index) și poate parsa structuri imbricate în tabele plate.

Mai jos este un mic exemplu de set de date JSON reprezentând un subset al Iris Dataset. Fiecare intrare este un obiect JSON cu câmpuri:

## 5 - Încărcarea din Baza de Date SQL


O altă sursă larg utilizată de seturi de date publice în fluxurile de lucru ML și analiză de date o reprezintă **bazele de date SQL**. Bazele de date SQL (Structured Query Language - Limbaj de Interogare Structurat) stochează date într-un format structurat, tabular, format din rânduri și coloane, organizate în tabele care pot fi legate între ele prin chei. Spre deosebire de fișierele plate precum `.csv` sau formatele ierarhice precum JSON, bazele de date SQL sunt proiectate pentru interogarea, filtrarea, agregarea și join-ul eficient al unor volume mari de date structurate.

Un set de date SQL este de obicei stocat într-o **bază de date**, care conține:
1. **Tabele** – colecții structurate de rânduri și coloane (similare DataFrame-urilor).
2. **Scheme** – grupuri logice de tabele și metadate asociate.
3. **Relații** – legături între tabele definite prin chei primare și chei externe.
4. **Indecși** – structuri de date care accelerează operațiunile de căutare și recuperare.

Deoarece bazele de date SQL impun scheme și tipuri de date, oferă garanții puternice privind consistența și integritatea datelor. Sunt utilizate frecvent în sistemele de producție, pipeline-uri de analiză și fluxuri de lucru ML la scară largă, unde seturile de date sunt prea mari sau prea dinamice pentru a fi stocate ca fișiere statice.

În bazele de date SQL, datele sunt interogate folosind instrucțiuni SQL precum:
```sql
SELECT sepal_length, sepal_width, species
FROM iris
WHERE species = 'setosa';
```
Această interogare recuperează un subset de rânduri și coloane, dintr-un tabel numit `iris`, care aparțin clasei țintă `setosa`. Astfel de interogări permit filtrarea și agregarea eficientă direct în motorul bazei de date înainte de a încărca rezultatele în Python.

Python oferă mai multe modalități de a interacționa cu seturile de date SQL:

1. Modulul built-in sqlite3 poate fi folosit pentru a executa interogări SQL față de bazele de date SQLite.
2. Conectorii specifici bazelor de date (ex.: psycopg2 pentru PostgreSQL, mysql-connector pentru MySQL) permit comunicarea cu bazele de date la distanță.
3. `pandas` oferă funcții de nivel înalt precum `pd.read_sql()`, `pd.read_sql_query()` și `pd.read_sql_table()` pentru a încărca rezultatele interogărilor SQL direct într-un `pd.DataFrame`.

Mai jos este un mic exemplu conceptual al unui tabel SQL reprezentând un subset al setului de date Iris:

| id  | sepal_length | sepal_width | petal_length | petal_width |  species  |
| :-: | :----------: | :---------: | :----------: | :---------: | :-------: |
|  1  |      5.1     |     3.5     |      1.4     |     0.2     |   setosa  |
|  2  |      4.9     |     3.0     |      1.4     |     0.2     |   setosa  |
|  3  |      6.2     |     3.4     |      5.4     |     2.3     | virginica |

Folosind `pd.read_sql()`, poți executa o interogare SQL și încărca setul de rezultate returnat direct într-un `pd.DataFrame`, combinând flexibilitatea interogărilor SQL cu comoditatea instrumentelor de manipulare a datelor pandas.

    ...

        sql (str | SQLAlchemy Selectable): Interogarea SQL de executat sau denumirea unui tabel.

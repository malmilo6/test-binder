### 1.3 Identifying Missing Values

Missing values (represented as `NaN` or `None`) are very common in real-world datasets. `pandas` provides `.isna()` to detect them:

````python
# Series with missing values
series_with_nulls: pd.Series = pd.Series(
    data=[1, np.nan, 3, None, 5, np.nan],
    name="WITH_MISSING"
)
series_with_nulls
````

````python
# Boolean mask: True where value is missing
print("Boolean mask (.isna()):")
print(series_with_nulls.isna())
````

````python
# Total count of missing values
print(f"Total missing values (.isna().sum()): {series_with_nulls.isna().sum()}")
````

````python
# Proportion of missing values
print(f"Proportion missing (.isna().mean()): {series_with_nulls.isna().mean()}")
````

### 3.2 Positional Selection with `.iloc[]`

`.iloc[]` selects data by **integer position** (zero-based index). It follows the same slicing rules as Python lists:

```py
df.iloc[row_indexer, col_indexer]
```

| Expression | Meaning |
|---|---|
| `df.iloc[0]` | First row |
| `df.iloc[:5]` | First 5 rows |
| `df.iloc[10:21]` | Rows at positions 10 to 20 (inclusive) |
| `df.iloc[-3:]` | Last 3 rows |
| `df.iloc[:, 0]` | First column (all rows) |
| `df.iloc[-3:, :2]` | Last 3 rows and first 2 columns |

<a name='33---loc-selection'></a>
### 3.3 Label-Based Selection with `.loc[]`

`.loc[]` selects data by **label** (row index label or column name). Unlike `.iloc[]`, when slicing with `.loc[]`, the stop label is **inclusive**:

```py
df.loc[row_label, col_label]
```

| Expression | Meaning |
|---|---|
| `df.loc[0, "sepalLength"]` | Single value at row label 0, column "sepalLength" |
| `df.loc[0:4]` | Rows with labels 0 to 4 (inclusive) |
| `df.loc[:, "sepalLength"]` | All rows, only "sepalLength" column |

````python
# Single value
val = sample_predef_df.loc[0, "sepalLength"]
print(f"Value at row 0, 'sepalLength' (df.loc[0, 'sepalLength']): {val}")
````

## 4 - GroupBy and Aggregation

Grouping data by categorical columns and then computing statistics per group is one of the most powerful analytical operations in `pandas`. The main method is `.groupby()`:

```py
df.groupby("group_column")["target_column"].aggregation_function()
```

After `.groupby()`, the result is a `DataFrameGroupBy` or `SeriesGroupBy` object. You then apply an aggregation method (`.mean()`, `.size()`, `.agg()`, etc.) to compute the statistic per group.

<a name='41---groupby'></a>
### 4.1 Grouping with `.groupby()`

`.groupby()` splits the DataFrame into groups based on unique values in one or more columns, then applies any aggregation method to each group:

````python
# Mean sepalLength per species
print("Mean sepalLength per species:")
print(sample_predef_df.groupby("species")["sepalLength"].mean())
````

````python
# Mean of ALL numeric columns per species
print("Mean of all numeric columns per species:")
print(sample_predef_df.groupby("species").mean(numeric_only=True))
````

<a name='42---group-sizes'></a>
### 4.2 Counting Group Sizes

`.size()` returns the number of rows in each group as a `pd.Series`:

````python
# Number of rows per species
print("Row count per species (.groupby().size()):")
print(sample_predef_df.groupby("species").size())
````

<a name='43---agg'></a>
### 4.3 Multiple Aggregations with `.agg()`

`.agg()` allows computing **multiple statistics** at once for a given column, returning a `pd.DataFrame` with one column per aggregation function:

```py
df.groupby("group_col")["target_col"].agg(["mean", "std", "min", "max"])
```

````python
# Multiple aggregations for sepalLength per species
print("sepalLength statistics per species (.agg()):")
sample_predef_df.groupby("species")["sepalLength"].agg(["mean", "std", "min", "max"])
````
